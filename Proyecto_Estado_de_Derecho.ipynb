# Exploratory Data Analysis (EDA)

```python
df = pd.read_excel('https://github.com/Claudio-Ferriz/Proyecto-Estado-de-Derecho-/raw/refs/heads/main/WJP_HISTORICAL_DATA_FILE(2025).xlsx')

df
```

```python
df.info()
```

```python
# Eliminar los valores correspondientes a 2012-2013

df["Year"] = pd.to_numeric(df["Year"])
```

## Visualizaciones

### Análisis Univariado

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.hist(df["WJP Rule of Law Index: Overall Score"].values.flatten(), bins=25)
plt.xlabel('Puntaje')
plt.ylabel('Cantidad')
plt.title('Histograma de Puntaje')
```

### Análisis Bivariado y Multivariado

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.hist(
    df.loc[df["Year"] == 2025, "WJP Rule of Law Index: Overall Score"], bins=25)

plt.xlabel('Puntaje')
plt.ylabel('Cantidad')
plt.title(f'Histograma de Puntaje – Año {2025}')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.hist(
    df.loc[df["Year"] == 2019, "WJP Rule of Law Index: Overall Score"], bins=25)

plt.xlabel('Puntaje')
plt.ylabel('Cantidad')
plt.title(f'Histograma de Puntaje – Año {2019}')
plt.show()
```

```python
plt.figure(figsize=[12,5])
plt.xticks(rotation=45)
sns.boxplot(df,x='Region',y='WJP Rule of Law Index: Overall Score')
```

```python
plt.figure(figsize=[12,5])
sns.scatterplot(data=df, x='Year', y='WJP Rule of Law Index: Overall Score', hue='Region')

plt.title('Puntaje por Región', fontsize=16)
plt.xlabel('Fecha', fontsize=12)
plt.ylabel('Puntaje', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title='País', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
